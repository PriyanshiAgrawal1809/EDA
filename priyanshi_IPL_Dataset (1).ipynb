```python
#Q-13 - WHAT IS THE PROBABLITY OF WINNING A MATCH IF THE TOSS WAS WON?
total_toss_won = df['toss_winner'].count()
print(total_toss_won)
# Count the number of matches won after winning the toss
matches_won_after_toss_won = df[(df['toss_winner'] == df['winner'])].shape[0]
print(matches_won_after_toss_won)
# Calculate the probability
probability = (matches_won_after_toss_won / total_toss_won) * 100
print("Probability of Winning after Winning Toss:", probability)
```
